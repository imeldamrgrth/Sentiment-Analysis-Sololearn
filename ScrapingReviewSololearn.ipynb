# Scraping Review dari PlayStore

### Mengambil lebih dari 10.000 review dari aplikasi Sololearn di Google Play.


```python
!pip install google-play-scraper
```

```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time

all_reviews = []
count = 0

while len(all_reviews) < 20000:
    new_reviews, _ = reviews(
        'com.sololearn',
        lang='id',
        country='id',
        count=200,
        sort=Sort.NEWEST,
        filter_score_with=None
    )
    all_reviews.extend(new_reviews)
    count += 1
    print(f"Sudah ambil {len(all_reviews)} review... [Batch {count}]")
    time.sleep(1)

# Simpan ke DataFrame
df = pd.DataFrame(all_reviews)
df = df[['content', 'score']]
df.to_csv('sololearn_raw.csv', index=False)
print(f"\nTotal review terkumpul: {len(df)}")
```
